```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(data=acq_data, x='cue1', y='cue5', hue='Group', palette='viridis', s=100)
plt.title('cue1 vs cue5 Distribution (Groups)')
plt.xlabel('cue1 Value')
plt.ylabel('cue5 Value')
plt.legend(title='Group')
plt.grid(True)
plt.show()
```

```python
acq_long = pd.melt(acq_data, id_vars=['Group'], value_vars=['cue1', 'cue2', 'cue3', 'cue4', 'cue5'],
                  var_name='Cue', value_name='Freezing')

plt.figure(figsize=(12, 6))
sns.scatterplot(data=acq_long, x='Cue', y='Freezing', hue='Group', palette='Set2')
plt.title('Acquisition Freezing')
plt.xlabel('Cue')
plt.ylabel('Freezing')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

acq_GLM =  smf.glm("Freezing ~ Group", data=acq_long, family=sm.families.Gaussian()).fit()

acq_GLM.summary()
```
